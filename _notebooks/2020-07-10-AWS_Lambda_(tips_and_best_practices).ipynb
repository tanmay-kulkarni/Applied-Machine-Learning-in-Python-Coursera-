## Accessing public internet using AWS Lambda

When you enable VPC access to a function, that function is able to access only the resources in that VPC.

If the lambda function needs access to both resources within the VPC as well as public internet, then the VPC also needs a NAT instance in it.

If the lambda function needs access to the internet, do NOT attach it to a public subnet or a private subnet without internet access.

Instead, attach it to a private subnet with internet access through a NAT instance or an Amazon VPC NAT Gateway.
